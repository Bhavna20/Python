```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from pandas_datareader import data
from scipy import stats
```

```python
#Download facebook and S&P 500 prices
fb=data.DataReader("FB", data_source="yahoo", start="2018-01-01" , end="2018-01-31")
SP=data.DataReader("^GSPC", data_source="yahoo", start="2018-01-01" , end="2018-01-31")
#pick the closing price for each facebook and S&P 500. Rename the columns accordingly
prices= pd.concat([fb['Close'], SP['Close']], axis=1)
prices.columns= ['FB', 'S&P']
```

```python
#  calculate the returns on each
monthly_return=prices.pct_change(1)
monthly_return.dropna(axis=0,inplace=True) #drop the first nan value
```

```python
#setting up regression model
y=monthly_return['FB'] 
x=monthly_return['S&P'] 
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

print('beta',slope) #beta
```
